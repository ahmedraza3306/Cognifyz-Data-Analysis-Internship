```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(
    style='whitegrid',
    context='talk'
)

df = pd.read_csv("../Dataset.csv")
```

```python
highest = df.nlargest(

    10,

    'Votes'

)

highest[
[
'Restaurant Name',

'Votes',

'Aggregate rating'

]
]
```

```python
lowest = df.nsmallest(

10,

'Votes'

)

lowest[
[
'Restaurant Name',

'Votes',

'Aggregate rating'

]
]
```

```python
corr = df[
[
'Votes',

'Aggregate rating'

]
].corr()

corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(13, 8.5), facecolor="#0B0F19")
ax.set_facecolor("#0B0F19")

sns.regplot(
    data=df,
    x='Votes',
    y='Aggregate rating',
    scatter_kws={
        'alpha': 0.3,
        'color': '#38BDF8',
        's': 30,
        'edgecolor': 'none',
        'zorder': 3
    },
    line_kws={
        'color': '#F43F5E',
        'linewidth': 2.5,
        'zorder': 4
    },
    ax=ax
)

ax.text(
    -0.06, 1.12, "User Engagement vs. Quality Ratings",
    transform=ax.transAxes, ha="left", va="center",
    fontsize=22, fontweight="bold", color="#FFFFFF"
)
ax.text(
    -0.06, 1.06, "Evaluating the correlation between total review votes and overall aggregate score.",
    transform=ax.transAxes, ha="left", va="center",
    fontsize=12, color="#94A3B8", style="italic"
)

ax.set_xlabel("Total Votes Received", fontsize=13, fontweight="bold", labelpad=15, color="#94A3B8")
ax.set_ylabel("Aggregate Rating (Out of 5.0)", fontsize=13, fontweight="bold", labelpad=15, color="#94A3B8")

plt.xticks(fontsize=11, color="#64748B")
plt.yticks(fontsize=11, color="#64748B")

for spine in ax.spines.values():
    spine.set_visible(False)

ax.grid(axis="both", linestyle="-", linewidth=0.8, color="#1E293B", alpha=0.6, zorder=0)
ax.tick_params(axis='both', which='both', length=0)

plt.text(
    0.02, 0.02, "Source: Zomato Restaurant Dataset", 
    transform=fig.transFigure, fontsize=10, style="italic", color="#475569"
)

plt.tight_layout()
plt.show()
```

# Key Insights

• Highly voted restaurants generally maintain better ratings.

• A moderate positive relationship exists between votes and ratings.

• Popular restaurants attract greater customer engagement.
