# Level 3 - Task 3 : Price Range vs Online Delivery and Table Booking

## Objectives

• Analyze relationship between restaurant price range and online delivery.

• Analyze relationship between restaurant price range and table booking.

• Determine whether premium restaurants provide additional services.

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(

    style='whitegrid',

    context='talk'

)

df = pd.read_csv("../Dataset.csv")
```

```python
delivery = pd.crosstab(

    df['Price range'],

    df['Has Online delivery']

)

delivery
```

```python
booking = pd.crosstab(

    df['Price range'],

    df['Has Table booking']

)

booking
```

```python
plt.figure(

    figsize=(12,7)

)

delivery.plot(

    kind='bar',

    stacked=True,

    colormap='viridis'

)

plt.title(

    'Price Range vs Online Delivery',

    fontsize=20,

    fontweight='bold'

)

plt.xlabel(

    'Price Range'

)

plt.ylabel(

    'Restaurants'

)

plt.xticks(

    rotation=0

)

plt.tight_layout()

plt.show()
```

```python
plt.figure(

    figsize=(12,7)

)

booking.plot(

    kind='bar',

    stacked=True,

    colormap='magma'

)

plt.title(

    'Price Range vs Table Booking',

    fontsize=20,

    fontweight='bold'

)

plt.xlabel(

    'Price Range'

)

plt.ylabel(

    'Restaurants'

)

plt.xticks(

    rotation=0

)

plt.tight_layout()

plt.show()
```

```python
delivery_percent = (

    delivery.div(

        delivery.sum(axis=1),

        axis=0

    )

    *100

)

delivery_percent.round(2)
```

```python
booking_percent = (

    booking.div(

        booking.sum(axis=1),

        axis=0

    )

    *100

)

booking_percent.round(2)
```

# Key Insights

• Higher-priced restaurants tend to offer table booking facilities.

• Online delivery is predominantly offered by low and mid-priced restaurants.

• Premium dining establishments focus more on dine-in experiences.

• Service availability changes noticeably across different price categories.

# Conclusion

Restaurant pricing significantly influences service offerings.

Lower and mid-range restaurants emphasize online delivery, while high-end restaurants are more likely to provide table booking facilities, enhancing customer dining experiences.
