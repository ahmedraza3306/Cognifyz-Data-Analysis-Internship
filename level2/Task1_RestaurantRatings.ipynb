```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(
    style='whitegrid',
    context='talk'
)

df = pd.read_csv("../Dataset.csv")

print("Dataset Loaded Successfully")

print("Rows :", df.shape[0])
print("Columns :", df.shape[1])

df.head()
```

## Dataset Overview

```python
df.info()
```

```python
df['Aggregate rating'].describe()
```

## Average Customer Votes

```python
avg_votes = df['Votes'].mean()

print(
    f"Average Number of Votes : {avg_votes:.2f}"
)
```

## Rating Range Analysis

```python
rating_bins = pd.cut(

    df['Aggregate rating'],

    bins=[0,1,2,3,4,5],

    labels=[

        '0-1',

        '1-2',

        '2-3',

        '3-4',

        '4-5'

    ]

)

distribution = (

    rating_bins

    .value_counts()

    .sort_index()

)

distribution
```

# Distribution of Restaurant Ratings

```python
plt.figure(

    figsize=(12,6)

)

colors = sns.color_palette(

    "magma",

    len(distribution)

)

bars = plt.bar(

    distribution.index,

    distribution.values,

    color=colors,

    edgecolor='black',

    linewidth=1.5

)

for bar in bars:

    height = bar.get_height()

    plt.text(

        bar.get_x()

        + bar.get_width()/2,

        height+50,

        f"{int(height):,}",

        ha='center',

        fontsize=11,

        fontweight='bold'

    )

plt.title(

    "Restaurant Rating Distribution",

    fontsize=20,

    fontweight='bold'

)

plt.xlabel(

    "Rating Range",

    fontsize=13

)

plt.ylabel(

    "Number of Restaurants",

    fontsize=13

)

plt.grid(

    axis='y',

    alpha=0.25

)

plt.tight_layout()

plt.show()
```

# Aggregate Rating Histogram

```python
plt.figure(

    figsize=(12,6)

)

sns.histplot(

    df['Aggregate rating'],

    bins=25,

    kde=True,

    color='#2563EB'

)

plt.title(

    'Aggregate Rating Distribution',

    fontsize=20,

    fontweight='bold'

)

plt.xlabel(

    'Aggregate Rating'

)

plt.ylabel(

    'Frequency'

)

plt.tight_layout()

plt.show()
```

# Key Insights

### Observations

• Restaurants are highly concentrated in the 3–5 rating range.

• Higher ratings generally indicate better customer satisfaction.

• Average votes help estimate restaurant popularity.

• Customer engagement appears stronger among well-rated restaurants.

# Conclusion

This analysis explored the distribution of aggregate restaurant ratings.

Most restaurants receive ratings above average, indicating generally positive customer experiences.
