# Level 1 - Task 2 : City Analysis

## Objective
This analysis aims to:

- Identify the city with the highest number of restaurants.
- Calculate the average restaurant rating for each city.
- Determine the city with the highest average rating.
- Visualize restaurant distribution and ratings across cities.

---
### Tools Used

- Python
- Pandas
- Matplotlib
- Seaborn

```python
# Import Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Professional visualization theme
sns.set_theme(style="whitegrid", context="talk")

# Load Dataset
df = pd.read_csv("Dataset.csv")

print("Dataset Loaded Successfully")
print(f"Rows : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

df.head()
```

```python
# Dataset Information

print("Dataset Information\n")

print(df.info())

print("\nMissing Values\n")

print(df.isnull().sum())
```

```python
# Restaurant count in each city

city_counts = (
    df["City"]
    .value_counts()
)

top10_city = city_counts.head(10)

print("Top 10 Cities by Restaurant Count\n")

top10_city
```

```python
# City with highest number of restaurants

highest_city = city_counts.idxmax()
highest_count = city_counts.max()

print(f"Highest Restaurant Count : {highest_city}")
print(f"Total Restaurants : {highest_count}")
```

```python
# Average Rating of Each City

average_rating = (
    df.groupby("City")["Aggregate rating"]
      .mean()
      .sort_values(ascending=False)
)

top10_rating = average_rating.head(10)

top10_rating
```

```python
highest_rating_city = average_rating.idxmax()
highest_rating = average_rating.max()

print(f"Highest Average Rating City : {highest_rating_city}")

print(f"Average Rating : {highest_rating:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(15, 8.5), facecolor="#F8FAFC")
ax.set_facecolor("#F8FAFC")

# 2. Define the exact colors for each specific bar
custom_colors = [
    "#3B82F6",  # New Delhi (Blue)
    "#0D9488",  # Gurgaon (Teal)
    "#F59E0B",  # Noida (Orange)
    "#6366F1",  # Faridabad (Purple)
    "#EF4444",  # Ghaziabad (Red)
    "#0EA5E9",  # Ahmedabad (Light Blue)
    "#D97706",  # Amritsar (Dark Gold)
    "#BE185D",  # Bhubaneshwar (Pink)
    "#0F766E",  # Guwahati (Dark Teal)
    "#78350F"   # Lucknow (Brown)
]

# 3. Create the bar chart (with curved tops approximated by avoiding harsh borders)
bars = ax.bar(
    top10_city.index,
    top10_city.values,
    color=custom_colors,
    width=0.75,
    zorder=3
)

# 4. Add the value labels above bars inside stylized dark pill tags
for bar in bars:
    height = bar.get_height()
    ax.annotate(
        f"{int(height):,}",
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 12),  # Offset text vertically
        textcoords="offset points",
        ha="center",
        va="center",
        fontsize=11,
        fontweight="bold",
        color="white",
        bbox=dict(
            boxstyle="round,pad=0.5", 
            facecolor="#1E293B", 
            edgecolor="none", 
            lw=0
        ),
        zorder=4
    )

# 5. Replicate the Header Title Design with top accents
ax.text(
    0.5, 1.08, "— — —   Top 10 Cities by Number of Restaurants   — — —",
    transform=ax.transAxes, ha="center", va="center",
    fontsize=24, fontweight="bold", color="#0F172A"
)
ax.text(
    0.5, 1.03, "Based on the Zomato Restaurant Dataset",
    transform=ax.transAxes, ha="center", va="center",
    fontsize=13, style="italic", color="#475569"
)

# 6. Replicate the Floating "Key Insight" Panel
insight_text = (
    "Key Insight\n\n"
    "New Delhi leads with the highest\n"
    "number of restaurants, followed by\n"
    "Gurgaon and Noida."
)
ax.text(
    0.96, 0.85, insight_text,
    transform=ax.transAxes, ha="right", va="top",
    fontsize=12, color="#1E293B", weight="medium",
    bbox=dict(
        boxstyle="round,pad=1.5", 
        facecolor="white", 
        edgecolor="#BFDBFE", 
        lw=1.5
    ),
    zorder=5
)

# 7. Axis Labels & Tick Styling
ax.set_xlabel("City", fontsize=16, fontweight="bold", labelpad=15, color="#0F172A")
ax.set_ylabel("Number of Restaurants", fontsize=16, fontweight="bold", labelpad=15, color="#0F172A")

plt.xticks(rotation=35, ha="right", fontsize=12, fontweight="bold", color="#1E293B")
plt.yticks(range(0, 6001, 1000), fontsize=13, fontweight="bold", color="#1E293B")

# Limit axes dynamically to accommodate the labels/titles comfortably
ax.set_ylim(0, 6000)

# 8. Clean up chart borders (Spines) to match the minimalist look
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(1.5)
ax.spines["left"].set_color("#0F172A")
ax.spines["bottom"].set_linewidth(1.5)
ax.spines["bottom"].set_color("#0F172A")

# 9. Dotted Gridlines
ax.grid(axis="y", linestyle="--", alpha=0.7, color="#CBD5E1", zorder=0)

# 10. Footer attribution
plt.text(
    0.02, 0.02, "Source: Zomato Restaurant Dataset", 
    transform=fig.transFigure, fontsize=10, style="italic", color="#64748B"
)

plt.tight_layout()
plt.show()
```
